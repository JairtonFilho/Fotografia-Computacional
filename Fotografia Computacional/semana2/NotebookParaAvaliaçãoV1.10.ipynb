```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
img = cv2.imread("C:\\Users\\Jairt\\Documents\\GitHub\\Fotografia-Computacional\\imgcourse\\lena.png")

gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img, cmap='gray')
plt.title('Lena')
plt.show()
```

```python
equalized_img = cv2.equalizeHist(gray_img)
```

```python
mean_value, std_dev = cv2.meanStdDev(equalized_img)
print(f"Média: {mean_value[0][0]:.2f}")
print(f"Desvio Padrão: {std_dev[0][0]:.2f}")
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(gray_img, cv2.COLOR_BGR2RGB), cmap='gray')
plt.title('Imagem Original em Tons de Cinza')

plt.subplot(1, 2, 2)
plt.imshow(equalized_img, cmap='gray')
plt.title('Imagem Equalizada')

plt.show()
```
